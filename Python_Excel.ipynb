```python
data = [["Mark",55,"Italy",4.5,"Europe"],
        ["John",33,"USA",6.7,"America"],
        ["Tim",41,"USA",3.9,"America"],
        ["Jenny",12,"Germany",9.0,"Eurpope"]

]
df = pd.DataFrame(data=data,
                  columns = ["Name","Age","Country","Score","Continent"],
                  index = [1001,1002,1003,1004])
df.info() #DataFrame的数据类型
df.dtypes #DataFrame列的数据类型
df.index #获取index对象
df.index.name = "Use_ID" #给index命名
df.reset_index() #重设index，将user的index还原为一般列，加入默认index
df.reset_index().set_index('Name') #链式调用，将names设定为新的index
```
